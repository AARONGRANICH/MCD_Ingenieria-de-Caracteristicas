# **Generalidad del proyecto**

Hacer un tablero que cuente una historia a partir de un análisis exploratorio de datos es un resultado parcial importante en un proceso de ciencia de datos. Estos tableros permiten ampliar la discusión y el entendimiento con los usuarios de los productos de datos y ayudan a plantear mejor las preguntas, seleccionar mejor los modelos de estimación y/o pronóstico así como a especificar más claramente los productos a desarrollar.

**Puntos a realizar:**

* Un archivo en markdown, en el cual se especifique:
La historia que queremos contar con nuestros datos.
Los KPI y otros indicadores que son útiles para contar la historia.
Al menos de un KPI, desarrollar la metodología tal como se muestra en alguno de los formatos facilitados
Un borrador (puede ser una foto tomada de un cuaderno) de como pensamos hacer el tablero de visualización.

# **Iniciamos con el proyecto**

### **¿ Qué historia queremos contar?**

Los datos que obtuvimos del INEGI (CENSO DE POBLACIÓN Y VIVIENDA 2020) y los datos de DATAMEXICO, los queremos utilizar para realizar un DASHBOARD donde nos muestre la situación de los Jóvenes en Sonora 2020 por municipio,  para con estos datos poder trabajar en políticas públicas, algunos de los datos que nos interesa mostrar son: 

-	Población total de 12 a 29 años
-	Población total 12 a 29 años Femenina 
-	Población total 12 a 29 años Masculina 
-	Población joven Escolarizada, ocupada y desocupada
-	Población joven que asiste a la escuela
-	Población joven Económicamente Activa
-	Población joven Analfabeta
-	Población joven con discapacidad
-	Población joven que habla una lengua indígena 
-	Población joven afroamericana o afrodescendiente 

Vamos a utilizar otras fuentes para hacer un mapa donde veamos donde se concentran las oportunidades de Educación, Empleos, Servicios de Salud, entre otros indicadores importantes, algunas de las fuentes son:
-	Anuarios Estadísticos de Educación Superior
-	Censo Población y Vivienda (Cuestionario ampliado).
-	Encuesta Nacional de Ocupación y Empleo (ENOE).
-	Data México

Con esta información queremos ayudar al ISJ (Instituto Sonorense de la Juventud) a destacar la importancia de diagnósticos precisos sustentados en datos, los cuales nos ayuden a tener un mejor diseño de estrategias, accione y políticas públicas con perspectiva juvenil, por lo tanto tener una mejor toma de decisiones para abordar la problemáticas en los distintos municipios del estado de Sonora.

## **¿Qué KPI (Key performance indicator / indicadores clave de desempeño)  vamos queremos obtener de estos datos)**

Algunos de los indicadores que podemos obtener con estos datos serían los siguientes:

-	Lugares donde se concentra la mayor población Joven 
-	Zonas con más oportunidades para la población Joven
-	Porcentajes de población joven con Discapacidad, Analfabetas, Hablante de lengua indígena, etc,
-	Porcentaje de escolarizad en los jóvenes
